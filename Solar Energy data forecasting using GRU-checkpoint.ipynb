```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
#from tensorflow.keras import Sequential
#from tensorflow.keras.layers import Conv2D, Flatten, Dense, Dropout, LSTM, GRU, Bidirectional, Flatten, TimeDistributed, RepeatVector
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, GRU, Bidirectional, Flatten, TimeDistributed, RepeatVector
#from tensorflow.keras.layers import Conv1D, MaxPooling1D
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.optimizers import SGD
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
#from tensorflow.keras.callbacks import History
from keras.callbacks import History
import math
import warnings
warnings.filterwarnings("ignore")
```

```python
# function for plot

def plot_predictions(test, predicted):
    plt.plot(test, color="red", label="real solar energy")
    plt.plot(predicted, color="blue", label="predicted solar energy")
    plt.title("  Solar Energy Forecasting  ")
    plt.xlabel("time")
    plt.ylabel("kWh")
    plt.legend()
    plt.show()
```

```python
# Function for Root_Mean_Squared_Error(RMSE)

def return_rmse(test, predicted):
    rmse = math.sqrt(mean_squared_error(test,predicted))
    print("the root mean squared error is : {}.".format(rmse))
```

```python
data = pd.read_csv("solar-energy-data.csv", index_col='timestamp', parse_dates=['timestamp'])
data.shape
```

```python
data.head(10)
```

```python
train = data[:'2013'].iloc[:,:1].values
#print(train)

test = data['2014':].iloc[:,:1].values
#print(test)
#test.shape
#train.shape
```

```python
# visualization of "kwh" attribute of the dataset

data[:'2013'].plot(figsize=(16,4), legend=True)
plt.legend(["Training set (before 2014)"])
plt.title("Solar Energy in kWh")
plt.ylabel("kWh")
data['2014':].plot(figsize=(16,4), legend=True)
plt.legend(["Test set (from 2014)"])
plt.title("Solar Energy in kWh")
plt.ylabel("kWh")
plt.show()
```

```python
# visualization of "kWh" attribute of the dataset

data['kwh'][:'2013'].plot(figsize=(16,4), legend=True)
data['kwh']['2014':].plot(figsize=(16,4), legend=True)
plt.legend([" Training set (before 2014) ", " Test set (from 2014) "])
plt.title(" Solar Energy in kWh ")
plt.ylabel(" kWh ")
plt.xlabel(" Date ")
plt.show()

```

```python
# scaling the training set

sc = MinMaxScaler(feature_range=(0,1))
train_scaled = sc.fit_transform(train)
```

```python
# x_train, y_train 

x_train = []
y_train = []

for i in range(6,275):
    x_train.append(train_scaled[i-6:i, 0])
    y_train.append(train_scaled[i,0])

x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
x_train.shape
```

```python
# now the Test set to prepare in the similar way as training set

dataset_total = pd.concat((data['kwh'][:'2013'], data['kwh']['2014':]), axis=0)
print(dataset_total.shape)

inputs = dataset_total[len(dataset_total)-len(test)-6:].values
print(inputs.shape)

inputs = inputs.reshape(-1,1)
print(inputs.shape)

inputs = sc.transform(inputs)
print(inputs.shape)
```

```python
# preparing the x_test

x_test = []

for i in range(6,371):
    x_test.append(inputs[i-6:i, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

#### Gated Recurrent Units (GRU)

```python


# GRU architecture

regressorGRU = Sequential()




# the first GRU layer with Update gate

regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='relu'))
regressorGRU.add(Dropout(0.2))

# 2nd layer of GRU

regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='relu'))
regressorGRU.add(Dropout(0.2))


# 3rd layer of GRU

regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='relu'))
regressorGRU.add(Dropout(0.2))


# 4th layer of GRU

regressorGRU.add(GRU(units=50, activation='relu'))
regressorGRU.add(Dropout(0.2))

# the output layer of GRU

regressorGRU.add(Dense(units=1))

```

```python
# for 500 epochs

#from keras.callbacks import History
#from tensorflow.keras.callbacks import History


history = History()

# compiling the model

regressorGRU.compile(optimizer='adam', loss='mse')

# fitting the model

history = regressorGRU.fit(x_train, y_train, epochs=1000, batch_size=150, shuffle=True, callbacks=[history])
```

```python
print(history.history.keys())
```

```python
val_loss = history.history['loss']
xc = range(500)

plt.figure()
plt.plot(xc, val_loss)
plt.title(" loss in epochs ")
plt.ylabel(" loss ")
plt.xlabel(" epochs ")
plt.show()
```

```python
# predicting 

predicting_with_gru = regressorGRU.predict(x_test)
predicting_with_gru = sc.inverse_transform(predicting_with_gru)
```

```python
# plot the graph

plot_predictions(test, predicting_with_gru)
```

```python
# evaluate the model

return_rmse(test, predicting_with_gru)
```
